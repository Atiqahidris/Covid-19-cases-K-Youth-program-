```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cases_age.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
age_columns = ['abs_0_4', 'abs_5_11', 'abs_12_17', 'abs_18_29', 'abs_30_39', 'abs_40_49', 'abs_50_59', 'abs_60_69', 'abs_70_79', 'abs_80_+']
total_cases_by_age = df[age_columns].sum()
```

```python
print (total_cases_by_age)
```

```python
plt.figure(figsize=(10, 6))
total_cases_by_age.plot(kind='bar')
plt.title('COVID-19 Cases by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
max_cases_age_group = total_cases_by_age.idxmax()
print(f"The age group with the highest number of cases is: {max_cases_age_group}")
```

```python
min_cases_age_group = total_cases_by_age.idxmin()
print(f"The age group with the lowest number of cases is: {min_cases_age_group}")
```

```python
# Pie Chart: Visualize the distribution of cases across age groups for a specific week.
week_data = df[df['week'] == 'w24'].set_index('state').sum(numeric_only=True)[age_columns]
week_data.plot(kind='pie', figsize=(10, 10), autopct='%1.1f%%')
plt.title('Distribution of COVID-19 Cases Across Age Groups for Week 24')
plt.ylabel('') #Remove the default 'None' label
plt.show()
```

```python
#Bar Chart: Visualize the total number of COVID-19 cases for each state.
# This line creates a new DataFrame (numeric_df) by selecting only the columns that contain numeric data types, specifically integers (int) and floating-point numbers (float). 
numeric_df = df.select_dtypes(include=[int, float])

# Ensure 'state' are included for filtering purposes
numeric_df['state'] = df['state']

statewise_cases = numeric_df[numeric_df['state'] != 'Malaysia'].groupby('state').sum().sum(axis=1)
plt.figure(figsize=(14, 7))
statewise_cases.sort_values(ascending=False).plot(kind='bar')
plt.title('Total COVID-19 Cases by State')
plt.xlabel('State')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Visualize the proportion of COVID-19 cases for each state.
plt.figure(figsize=(10, 10))
statewise_cases.sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of COVID-19 Cases by State')
plt.ylabel('') # Remove the default 'None' label
plt.show()
```

```python
# Filter out only numeric columns
numeric_df = df.select_dtypes(include=[int, float])

# Ensure 'week' and 'state' are included for filtering purposes
numeric_df['week'] = df['week']
numeric_df['state'] = df['state']

# Filter the data for the specific state
specific_state = 'Johor'
state_data = numeric_df[numeric_df['state'] == specific_state]

# Check if there is data for the specific state
if state_data.empty:
    print(f"No data available for {specific_state}")
else:
    # Set 'week' as index and sum across numeric columns
    state_data = state_data.set_index('week').drop(columns=['state']).sum(axis=1)

    # Plotting
    plt.figure(figsize=(12, 6))
    state_data.plot()
    plt.title(f'Trend of COVID-19 Cases in {specific_state} Over Weeks')
    plt.xlabel('Week')
    plt.ylabel('Number of Cases')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
